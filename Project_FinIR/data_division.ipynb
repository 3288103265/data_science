```python
def extract_opinion(string, source, metal_relevant):
    if source == 'https://www.economies.com':
        keyword = ['expect', ' stochastic ', ' will ', ' if ', ' we '] + metal_relevant
        split_string = re.split('\. |\.\\xa0|\\xa0', string)
        res = [i for i in split_string if len([j for j in keyword if j in i.lower()])>0]
    elif source == 'https://www.fxempire.com':
        keyword = [' trend ', ' will ', ' indicate ', ' if' , ' may ', ' likely ', ' could ', ' we ']+ metal_relevant
        split_string = re.split('\. |\.\\xa0|\\xa0', string)
        res = [i for i in split_string if len([j for j in keyword if j in i.lower()])>0]
    elif source == 'https://news.metal.com':
        keyword = [' will ', ' predict ', ' believe ', 'expect', 'due to', 'causing', 'trend', 'in summary', 'should', 'continue']+ metal_relevant
        split_string = re.split('\. |\.\\xa0|\\xa0', string)
        res = [i for i in split_string if len([j for j in keyword if j in i.lower()])>0]
        res += split_string[-3:-1]
        res = sorted(set(res), key=res.index)
    elif source == 'https://www.metalbulletin.com':
        keyword = ['due to', 'fastmarkets', 'forecast', ' mean ', 'may', 'should', 'will', 'could']+ metal_relevant
        split_string = re.split('\. |\.\\xa0|\\xa0', string)
        res = [i for i in split_string if len([j for j in keyword if j in i.lower()])>0]
    elif source == 'http://smconline.smcindiaonline.co.in/Commodity_Analysis/':
        keyword = [' likely ', 'expect', 'because', ' could ', ' can ', ' would ', ' if ', ' may ']+ metal_relevant
        split_string = re.split('\. |\.\\xa0|\\xa0', string)
        res = [i for i in split_string if len([j for j in keyword if j in i.lower()])>0]
        res += split_string[-2:]
        res = sorted(set(res), key=res.index)
    elif source == 'https://economictimes.indiatimes.com':
        keyword = ['analysts', 'due to', 'expect', 'may', 'trader', 'could', 'will', 'should']+ metal_relevant
        split_string = re.split('\. |\.\\xa0|\\xa0', string)
        res = [i for i in split_string if len([j for j in keyword if j in i.lower()])>0]
    return ';'.join(res)
```

```python
def select_contain_metal_sentence(string, metal_lst):
    divided_string = english_sent_tokenizer.tokenize(string)
    res = [i for i in divided_string if sum([j in i.lower() for j in metal_lst])!=0]
    return ';'.join(res)
```
